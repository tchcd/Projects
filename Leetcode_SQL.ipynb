**185. Department Top Three Salaries**

**Hard**

A company's executives are interested in seeing who earns the most money in each of the company's departments. A high earner in a department is an employee who has a salary in the top three unique salaries for that department.

Write an SQL query to find the employees who are high earners in each of the departments.

Return the result table in any order.


```python
Employee = pd.DataFrame({"id":[1,2,3,4,5,6,7], "name":["Joe", "Henry", 'Sam', 'Max', 'Janet', 'Randy', 'Will'], 
                        "salary":[85000, 80000, 60000, 90000, 69000, 85000, 70000], "departmentId":[1,2,2,1,1,1,1] })
```

```python
Department = pd.DataFrame({"id":[1,2], "name":['IT', 'Sales']})
```

```python
Employee.to_sql('Employee', con, index=False, if_exists = 'replace') 
Department.to_sql('Department', con, index=False, if_exists = 'replace')
```

```python
sql = '''
SELECT t.Department, t.Employee, t.salary FROM
(SELECT d.name AS Department, t.name AS Employee, t.salary, DENSE_RANK() OVER (partition by d.id ORDER BY t.salary DESC) as rnk FROM Employee t
INNER JOIN Department d on t.departmentId=d.id) t
WHERE t.rnk <=3
ORDER BY t.salary DESC
''' 
```

```python
select(sql)
```

```python
sql = '''
WITH rnk_dep_1 AS (
SELECT t.*, DENSE_RANK() OVER(ORDER BY t.salary DESC) AS rnk FROM Employee t
WHERE t.departmentId = 1),

rnk_dep_2 AS (
  SELECT t.*, DENSE_RANK() OVER (ORDER BY t.salary DESC) AS rnk FROM Employee t
  WHERE t.departmentId = 2
),

whole AS (
SELECT * FROM rnk_dep_1 d1
UNION 
SELECT * FROM rnk_dep_2 d2)

SELECT d.name as Department, w.name as Employee, w.salary FROM whole w

INNER JOIN Department d on w.departmentId = d.id

WHERE rnk <= 3
ORDER BY w.salary DESC
'''
```

**Medium**

Write an SQL query to rank the scores. The ranking should be calculated according to the following rules:

The scores should be ranked from the highest to the lowest.

If there is a tie between two scores, both should have the same ranking.

After a tie, the next ranking number should be the next consecutive integer value. In other words, there should be no holes between ranks.

Return the result table ordered by score in descending order.


```python
obj = {"headers": ["ID", "SCORE"], "values": [[1, 3.5], [2, 3.65], [3, 4.0], [4, 3.85], [5, 4.0], [6, 3.65]]}
```

```python
sql = '''
SELECT t.score, t.rnk FROM
(SELECT t.score, COUNT(DISTINCT s.score) as rnk FROM Scores t
JOIN scores s on t.score <= s.score
GROUP BY t.score
ORDER BY COUNT(DISTINCT s.score)) t
JOIN scores s on t.score = s.score
ORDER BY t.rnk
'''
```

Write an SQL query to find all numbers that appear at least three times consecutively.

```python
sql = '''
drop table if exists Logs;
Create table If Not Exists Logs (id int, num int);
insert into Logs (id, num) values ('1', '1');
insert into Logs (id, num) values ('2', '1');
insert into Logs (id, num) values ('3', '1');
insert into Logs (id, num) values ('4', '2');
insert into Logs (id, num) values ('5', '1');
insert into Logs (id, num) values ('6', '2');
insert into Logs (id, num) values ('7', '2');'''

cur.executescript(sql)
```

```python
sql = '''
SELECT CAST(t.ANS AS SIGNED) AS ConsecutiveNums FROM(
SELECT t.*,
CASE WHEN LEAD_1 = num AND LEAD_2 = num THEN t.num ELSE 'Null' END AS ANS
 FROM 
(
SELECT t.*,
LEAD(t.num, 1) OVER (ORDER BY t.id) AS LEAD_1,
LEAD(t.num, 2) OVER (ORDER BY t.id) AS LEAD_2
FROM Logs t) t) t
WHERE t.ANS not in ('Null')
GROUP BY t.ANS
'''
```

```python
sql = '''
WITH cums AS(
SELECT t.*,
SUM(t.flag) OVER (ORDER BY t.id) as cum_sum
 FROM (
SELECT t.*,
LAG(t.people) OVER (ORDER BY t.id) as prev,
CASE WHEN t.people >= 100 AND LAG(t.people) OVER (ORDER BY t.id) >= 100 THEN 0 ELSE 1 END AS flag
FROM Stadium t) t)

SELECT t.id, t.visit_date, t.people FROM cums t
WHERE t.cum_sum in
(SELECT t.cum_sum FROM cums t
GROUP BY t.cum_sum
HAVING count(1) >=3)
ORDER BY t.visit_date
'''
```

```python
sql = '''
WITH cte AS(
SELECT t.*,
SUM(t.flag) OVER (ORDER BY t.id) as cum_sum
 FROM (
SELECT t.*,
CASE WHEN t.people >= 100 AND t.people2 >= 100 THEN 0 ELSE 1 END AS flag
FROM (
SELECT t.*, s.people as people2 FROM Stadium t
LEFT JOIN Stadium s on t.id=s.id+1
) t ) t )

SELECT * FROM cte t
WHERE t.cum_sum in (
SELECT t.cum_sum FROM cte t
GROUP BY t.cum_sum
HAVING count(1) >=3)
'''
```

```python
sql = '''
WITH cte AS (
SELECT t.*,
CASE WHEN t.customerID != "NaN" THEN 1 ELSE 0 END AS flag
 FROM (
SELECT *FROM Customers t
LEFT JOIN Orders o on t.id=o.customerId
) t)

SELECT t.name as Customers FROM cte t
WHERE t.flag = 0

'''
```
